```python
from sage.plot.colors import rainbow
C = graphs.CubeGraph(3)
R = rainbow(3)
edge_colors = {}
for i in range(3):
    edge_colors[R[i]] = []
for u,v,l in C.edges():
    for i in range(3):
        if u[i] != v[i]:
            edge_colors[R[i]].append((u,v,l))
C.graphplot(vertex_labels=False, 
            vertex_size=0, edge_colors=edge_colors).show()
```
