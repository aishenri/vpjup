```python
x = var('x') 
a = 0.5 
# Halbachse a der Ellipse 
b = 0.75 
# Halbachse b der Ellipse 
f1 = b*sqrt(1-x^2/a^2) 
# obere Halbellipse 
f2 = -b*sqrt(1-x^2/a^2) 
# untere Halbellipse 
# ch der Ellipse [0,a] 
g = Graphics() 
g += plot(f1,xmax=a,xmin=-a,ymax=1,ymin=-1) 
g += plot(f1,xmax=a,xmin=0,
          ymax=1,ymin=-1,
          color='red',legend_label='Funktion y(x)')
g += plot(f2,xmax=a,xmin=-a,
          ymax=1,ymin=-1,linestyle='--') 
g += point((0,0),color='green',
           size=40, legend_label='front') 
g += point((a,0),color='green',size=40) 
g.show() 
```
