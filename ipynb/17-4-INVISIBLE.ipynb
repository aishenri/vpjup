##### Invisible

Un atome est invisible... A partir d'une certaine taille on ne voit plus les choses ? Pourquoi ?

Je pense que la réponse est une question de vitesse ?

Depuis Maxwell et avec Einstein, les scientifiques font une erreur extrêment grave : Ils parlent de vitesse de la lumière quand il s'agit de fréquence : La formule de la célérité $c$ n'est pas une vitesse mais une fréquence de résonance, en fait au carrée (une accélération) ! $c^2 = \frac {1}{\mu \epsilon}$

Pour comprendre le problème il suffit de prendre $\pi$

A l'extérieur du cercle $\pi$ vaut 3 : Un dizième vaut 3.1, intérieur du cercle, la réduction est elle qu'à un millième on ne voit plus rien, mais $\pi$ existe toujours (potentiellement), il est devenu imaginaire.


```python
c=circle((0,0),3);show(c,figsize=3)
```

```python
c1=circle((0,0),3/1.1,color="red");show(c1+c,figsize=3)
```

```python
c2=circle((0,0),3/1.15,color="green");show(c+c1+c2,figsize=3)
```

```python
c3=circle((0,0),3/1.2,color="violet");
show(c+c1+c2+c3,figsize=3,axes=False)
```
