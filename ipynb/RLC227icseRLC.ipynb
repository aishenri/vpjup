```python
def phi(x,m): 
    return 1/sqrt(2^m*factorial(m))*pi.n()^(-0.25)*exp(-x^2/2)*hermite(m,x) 
p=[] 
for i in range(8): 
    p.append(plot(phi(x,i)+i+0.5,(x,-5,5),rgbcolor=(1-0.125*i,0.5,0.12*i),thickness=1.5,legend_label="n="+str(i))) 
    p.append(plot(i+0.5,(x,-5,5), color="black",linestyle=":",thickness=1)) 
p.append(plot(0.5*x**2,(x,-5,5), color="black",linestyle="--",thickness=2)) 
sum(p).show(figsize=5,ymax=8,
            frame=True,fontsize=8,axes=False) 
```
