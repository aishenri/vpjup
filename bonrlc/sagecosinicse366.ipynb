```python
def Papier_Milimetrowy(dane,xtics=[],
                       ytics=[],xlab=[],
                       ylab=[]): 
    from matplotlib import ticker 
    papier=sum([line([[i,0],[i,4]],
                     thickness=0.51,
                     color='gray') 
for i in range(9)])+sum([line([[i/10,0],
                               [i/10,4]],
                              thickness=0.3,
                              color='gray') 
for i in range(81)])+sum([line([[0,i],[8,i]],
                               thickness=0.6,
                               color='gray') 
for i in range(5)])+sum([line([[0,i/10],
                               [8,i/10]],
                              axes='frame',
                              xmin=-0.1,
                              xmax=8.1,
                              ymin=-0.1,
                              ymax=4.1,
                              figsize=(8,4),
                              thickness=0.3,
                              color='gray') 
for i in range(41)]) 
    plt=point(dane,color='red',size=30)+papier
    plt.show( ticks=(xtics,ytics),
             tick_formatter=( ticker.FixedFormatter(xlab),
                             ticker.FixedFormatter(ylab) )) 
```
