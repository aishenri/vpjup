```python
import pymol;from pymol import cmd
pymol.pymol_argv = ['pymol', '-qc']
from IPython.display import Image
cmd.load('1ala.pdb')
!pwd
```

```python
cmd.do('reinitialize;load glycin.pdb;show cartoon;orient;center;\
preset.ball_and_stick(selection="all", mode=1);\
show sticks, resn GTP;show spheres, name MG;\
color yellow, chain A;\
set label_size,12;rotate x,180;\
color black, elem C;viewport 300,300;')
cmd.do('label (gly),name')
cmd.png('glycin.png', ray=1);Image(filename='glycin.png')
#from IPython.display import Image
#cmd.png('1crn.png', ray=1);Image(filename='1crn.png')
```
