```python
var("x")
y1 = function("y1")(x)
y2 = function("y2")(x)
ec1=diff(y1, x) == -y2
ec2 = diff(y2, x) == y1
desolve_system([ec1 , ec2], [y1 , y2], 
                ics=[0, 1,0])
```

```python
[solx , soly] = desolve_system ([ec1 , ec2], 
                                [y1 , y2], 
                                ics=[0, 1, 0])
solx = solx.rhs()
soly=soly.rhs()
fig1 = plot(solx , 0, 2*pi , color="blue",
            legend_label="$cos(x)$")
fig2 = plot(soly , 0, 2*pi , color="red", 
            legend_label="$sin(x)$")
show(fig1 + fig2,figsize=4,axes=False,
     frame=True)
```
