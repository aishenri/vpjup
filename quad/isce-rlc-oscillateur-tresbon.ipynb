```python
%display latex
t = var('t')
var('L, C, t0, q0, I0')
q = function('q')(t)
assume(L>0)
assume(C>0)
assume(t>0)
LC = L*diff(q,t,2)+q/C == 0;LC
```

```python
q_LCo = desolve(LC, q, ivar=t);q_LCo
```

```python
q_LC = desolve(LC, q, ivar=t, ics=[t0,q0,I0]);q_LC
```

```python
I=diff(q_LC, t, 1);I
```

```python
plotq_LC=q_LC.substitute(t0=0, q0=1, I0=0, L=12e-2, C=1.5e-4);
```

```python
p_c=plot(plotq_LC, (0, 0.05),color="red")
```

```python
plot_I=I.substitute(t0=0, q0=0.01, I0=0, L=12e-1, C=1.5e-5);
```

```python
p_i=plot(plot_I, (0, 0.05), color='blue')
show(p_c+p_i)
```
