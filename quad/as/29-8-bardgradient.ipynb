```python
f(x,y) = cos(x)-2*sin(y)
gradient = derivative( f )
```

```python
nab=var("nabla",latex_name=r'\mathcal{\nabla f}')
```

```python
nabla=gradient;show(f);show(nab,nabla)
```

```python
P1 = plot_vector_field(gradient, (x,-10,10), (y,-10,10));
show(P1,figsize=3,fontsize=6)
```

```python
P2 = contour_plot(f, (x,-10,10), (y,-10,10), fill=False)
show(P2,figsize=3,fontsize=6)
```

```python
show(P1+P2,figsize=5,fontsize=6)
```

```python
var("y")
x1=0;y1=0;x2=1.9;y2=0
r_vec_1 = vector( [ x1-x, y1-y ] )
r_vec_2 = vector( [ x2-x, y2-y ] )
q0= -1;q1=2;q2= 1
f1 = - q0*q1*r_vec_1/r_vec_1.norm()^3
f2 = - q0*q2*r_vec_2/r_vec_2.norm()^3
net_f = f1 + f2
plot_vector_field(net_f, (x, 0.01, 1.9),
(y, -2, 2), plot_points=30,figsize=4,fontsize=6,axes=False)
```

```python
E.<x,y> = EuclideanSpace()
E.default_frame()
```

```python
v = E.vector_field(cos(x),-2*sin(y),name='v')
v.display()
```

```python
v.plot()
```

```python
p = E((-2,3), name='p');p.coord()
```

```python
vp = v.at(p);vp.display()
```
