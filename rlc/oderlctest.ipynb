```python
var("Q,f,omega,t")
```

```python
y=function('y')(t)
eq=diff(y,t,2)+omega*diff(y,t)+omega^2*y\
==cos(omega*t)+sin(omega*t);eq
```

```python
omega=2*pi*1000;Q=5;f=100;
eq=diff(y,t,2)+omega*diff(y,t)/Q+omega^2*y\
==sin(10*pi*f)
```

```python
sol1=desolve(eq,y,ics=[0,1,1],show_method=True)
```

```python
f1=plot(sol1[0],t,0,1/500,
        color='purple',thickness=2);
show(f1,gridlines='major',
     axes_labels=["$t$","$s(t)$"],
     frame=True,fontsize=14,axes=False)
```
