```python
def dist(u,v):
    h = [(u[i]+v[i])%2 for i in range(len(u))]
    return sum(h)
```

```python
u=(1,0,1,1,1,0)
v=(0,0,1,1,0,0)
dist(u,v)
```

```python
def cube(n):
    G = Graph(2**n)
    vertices = Tuples([0,1],n)
    for i in range(2**n):
        for j in range(2**n):
            if dist(vertices[i],vertices[j]) == 1:
                 G.add_edge(i,j)
    return G
```

```python
cube(3).plot3d(viewer="threejs")
```

```python
cube(4).plot3d(viewer="threejs")
```

```python
G = cube(2)
G.adjacency_matrix().eigenvalues()
```
