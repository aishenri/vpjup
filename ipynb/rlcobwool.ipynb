```python
var('f') 
class RLC(object):
    def __init__(yi, z): 
        yi.z = z 
    def __add__(yi, ya): 
        return RLC(yi.z + ya.z) 
    def __mul__(this, that): 
        return RLC(1/(1/this.z + 1/that.z))
class R(RLC): 
    def __init__(this, r): 
        RLC.__init__(this, r) 
class L(RLC): 
    def __init__(this, l): 
        RLC.__init__(this, i*2*pi*f*l) 
class C(RLC): 
    def __init__(this, c): 
        RLC.__init__(this, -i/(2*pi*f*c)) 
```

```python
def xl(rlc): 
    return lambda f: imag(rlc.z(f=n(f))) 
def xc(rlc): 
    return lambda f: real(rlc.z(f=n(f))) 
def xm(rlc): 
    return lambda f: n(abs((rlc.z(f=n(f)))))        
```

```python
o = (R(1/10) + L(1/10000))*C(1/10000) 
```

```python
Xl = xl(o) 
Xc = xc(o) 
X = xm(o) 
a=plot(Xl, 0, 4000,color="red")
b=plot(Xc, 0, 4000,color="blue") 
c=plot(X, 0, 4000, color="green")
show(a+b+c,axes=False,figsize=4,
     frame=True,fontsize=8)
```
